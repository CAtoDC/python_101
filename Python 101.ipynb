### Variables and Pass-By-Reference
When you assign a variable in Python, you are creating a reference to the object. In some languages an assignment from one variable to another would cause the data to be copied. In Python, the variables refer to the same object.

Variables can be of any datatype - and are mutable.

```python
a = [1, 2, 3]  # create a list with values

b = a          # Assign a to b

print('a = ', a)       # a = 1, 2, 3
print ('b = ', b)      # b = 1, 2, 3

a.append(4)    # append the number 4 to a; a now equals 1, 2, 3, 4

print ("a = ", a)
print ("b = ", b)  # b refers to the same object as a, so b also equals "1, 2, 3, 4"
```

**Averaging** with a list

```python
# this gets input of numbers from the user
# when the user is done entering numbers, he types:  done
# the system then averages the numbers entered (as a float)

numlist = list()

while True:
    inp = input("Enter a number: ")
    
    # We're done entering numbers when we type 'done'
    if inp == 'done':
        break
    
    our_sum = float(inp)
    
    # This gets added to with every iteration
    numlist.append(our_sum)
    
average = sum(numlist)/len(numlist)

# formats the float to 2 decimal points
print ('The average is:', str("%.2f" % average) +'.')
```

### List comprehension (using **split**)<a id="listcomp"></a>

List comprehension is an elegant way to define and create list in Python. List comprehension is a complete substitute for the lambda function as well as the functions map(), filter() and reduce(). For most people the syntax of list comprehension is easier to be grasped. 
A list comprehension consists of brackets containing an expression followed by a **for** clause, then zero or more **for** or **if** clauses. 

```python
s = 'the quick brown fox jumped over the lazy dog'
words = s.split()

# For each word in words, if there is an 'o' in the word, put it in the variable "word"
# and make it uppercase. Continue until all words have been covered.
# List comprehension is very similar to a lambda function

[word.upper() for word in words if 'o' in word]
```

## Dictionaries<a id="dictionaries"></a>

Broadly, a Dictionary object is a "bag of variables", each with its own label. More specifically, a Dictionary (dict) is a hash map (in Perl or PHP it would be an Associative Array). It's a flexibly-sized collection of key/value pairs, where *key* is the label and *value* is the thing itself. Dictionaries are extremely fast - and are often used as look up tables.

The values in a Dictionary don't stay in order. Think **content of purse**.

**Lists** index their entries based on their place in the list.

**Dictionaries** index entries via a "lookup tag".

```python
purse = dict()
purse['dollars'] = 12
purse['candies'] = 10
purse['tissues'] = 3

# the values of the dict don't stay in order
print (purse)

print ("How many candies are in the purse?", purse['candies'])

purse['candies'] = purse['candies'] + 2

print ("We added candies. How many are there now?", purse['candies'])
```

It is an **error** to reference a key that is not in the dictionary.

```python
abc = dict()

# This will get an error
try:
    print (abc['mark'])
except KeyError:
    print('Error: Item is not in the dictionary.')
```

But we can find out whether a value is in the dict by using the **in** operator - essentially asking Python whether the value is **in** the dict

```python
# This will result in a boolean value
abc = dict()
print ('mark' in abc)
```

### Determining whether an object is iterable

```python
def isiterable(obj):
    try:
        iter(obj)
        return True    # iterable
    except TypeError:  # not iterable
        return False

# Test - strings are iterable; integers are not
print (isiterable("I'm a string!"), '- strings are iterable')
print (isiterable(5), '- integers are not')
```

### Convert a non-iterable object to an iterable object

```python
# Use this when writing functions that can accept multiple kinds of input. 

x = 5
# print (isiterable(x))

if not isinstance(x, list) and isiterable(x):
    x = list(x)
print ('x is iterable:', isiterable(x))
```

## Generators<a id="generators"></a>

Functions only get one chance to return results and therefore must return all results at once.

A generator generates values (plural) using the **yield** keyword.

Key takeaways:

* Generators are used to generate a series of values
* **`Yield`** is like the return of generator functions
* The only other thing yield does is save the "state" of a generator function
* A generator is just a special type of iterator
* Like iterators, we can get the next value from a generator using `next()`
    * **`For`** gets values by calling **`next()`** implicitly

```python
def simple_generator_function():
    yield 1
    yield 2
    yield 3
    
for value in simple_generator_function():
    print(value)
```

```python
# we can also do this
our_generator = simple_generator_function()
# yields the first one
next(our_generator)
```

```python
# yields the second one
next(our_generator)
```

```python
# yields the third one
next(our_generator)
```

## Counting Pattern

The general pattern to count the words in a line of text is to **split** the line into words, then loop through the words and use a **dictionary** to track the count of each word independently.

Here we list the words in the dictionary and count how many times they appear in a sentence.

```python
# this counts each word in a line
# and keeps track of how many times the
# word is used
counts = dict()
print ('Enter a line of text: ')
line = input('')

words = line.split()
print ('Words: ', words)

print ('Counting . . . ')

for word in words:
    counts[word] = counts.get(word, 0) + 1
print ('Counts: ', counts)
```

## Using pylab to define a parabola

The equation for a parabola is: $y = ax^2 + bx + c$

```python
# In iPython or the iPython notebook, it's easiest to use pylab magic (%pylab inline),
# which imports matplotlib, numpy, and scipy.

# The inline flag means that images will be shown here in the notebooks, rather
# than in pop-up windows.

%pylab inline

x = range(-10, 11)
print(x)

y = []

a = 2
b = 3
c = 8

for xi in x:
    yi = a * xi ** 2 + b * xi + c
    # print(xi, yi)
    y.append(yi)

pylab.plot(x,y)

grid()
xlabel('$x$')
ylabel('$y$')
title('My Parabola')
```

is used to print a human readable presentation of an object. In this case it prints the class name and some other attributes.

```python
class Point:
    def __init__(self, x, y):
        self.x, self.y = x, y
        
    def __repr__(self):
        return 'Point(x=%d, y=%d)' % (self.x, self.y)
    
p = Point(1, 2)
p
```

## Symbolic math using sympy

SymPy is a Python library for symbolic mathematics. It aims to become a full-featured computer algebra system (CAS) while keeping the code as simple as possible in order to be comprehensible and easily extensible. SymPy is written entirely in Python and does not require any external libraries.

```python
from sympy import *
init_printing(use_latex=True)
x = Symbol('x')
y = Symbol('y')
series(exp(x), x, 1, 5)
```

```python
eq = ((x+y)**2 * (x+1))
eq
```

```python
expand(eq)
```

### Elegant Fibonacci code

```python
# Fibonacci series
# The sum of two elements defines the next

a, b = 0, 1
while b < 100:
    print (b)
    # let b equal a and let a+b = b
    a, b = b, a + b
```

### Fibonacci using a Generator

```python
# Fibonacci series
# The sum of two elements defines the next
# Using a generator

def fib(a = 1, b = 1):
    while True:
        yield a
        a, b = b, a + b
        
from itertools import islice
list(islice(fib(), 11))
```
